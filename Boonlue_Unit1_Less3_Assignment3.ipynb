###1. Choosing one variable and plotting in four different ways

####1.1 Histogram

```python
plt.hist(df.Age, color='gold',  alpha=.5, bins=np.arange(10,80))
plt.xlabel('Age of Olympic Athletes')
plt.title("Frequency of Olympic Athletes'Age")

plt.show()

#It is easy to see the age distribution of atletes with a histogram, 
#can easily tell that the majority of the athletes are between 20 to mid 30s
```

####1.2 Kernel Density Estimation

```python
sns.distplot(df.Age, color='darkorchid', hist=False)
plt.xlabel('Age of Olympic Athletes')
plt.title("Frequency of Olympic Athletes'Age")
plt.show()

#It is relatively easy to see the age distribution with a kernel density
#estimation, but not as intuitive as the histogram. 
```

####1.3 Violin Plot

```python
sns.violinplot(y=df.Age)
plt.title("Age Distribution of Olympic Athletes")
plt.show()

#The violin plot allows for easy visualization of the distribution, but
#doesn't provide information on the number of frequency or how many 
#athletes are that age. 
```

####1.4 Box Plot

```python
ax = sns.boxplot(y=df.Age, color='forestgreen')  
plt.title('Age Distribution of Olympic Atletes')
plt.show()

#The box plot is inadequate for showing age because there are too many 
#outliers. As a result, the graph isn't easy on the eyes. The positive 
#aspects about this graph is that it shows the quantiles and the average.
```

###2. Choosing two continuous variables and graphing them in 3 different ways

####2.1 Scatter Plot

```python
sns.lmplot(x='Height',y='Weight', data=df,fit_reg=False, scatter_kws={'alpha':0.4}, palette='muted')
plt.xlabel("Height in Centimeters")
plt.ylabel('Weight in Kilograms')
plt.title('Scatterplot: Weight as a Function of Height for Olympic Athletes')
plt.show()

#Great to see how height and weight correlate with one another, but due 
#to the large data set, it is somewhat difficult to see the scatter of points.
```

####2.2 Kernel Density

```python
sns.jointplot(x='Height', y='Weight', data=df, kind="kde")
plt.xlabel("Height in Centimeters")
plt.ylabel('Weight in Kilograms')
plt.title('Kernel Density: Weight as a Function of Height for Olympic Athletes')
plt.show()

#The graph is able to show the weight and height of athletes that occur the most often 
#as well as show an intriguing graphic. I think this type of graph represent the 
#weight and height attributes in the data set well enough.
```

####2.3 Hexbin Plot

```python
sns.jointplot(x='Height', y='Weight', data=df, kind="hex", color='orange')
plt.xlabel("Height in Centimeters")
plt.ylabel('Weight in Kilograms')
plt.title('Kernel Density: Weight as a Function of Height for Olympic Athletes')
plt.show()

#
```

###3. Choosing one continuous variable and one categorical variable, then graphing them in 6 different ways

####3.1 Boxplot

```python
sns.set(style='whitegrid',palette='pastel')

ax=sns.boxplot(x=df.Sex, y=df.Height,hue=df.Season,palette=['r','c'])

plt.title('Height by Gender and Olympic Season')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Height in Centimeters')
plt.show()

#The graph makes it easy to differentiate between the height of male and female 
#athletes and also between summer vs. winter Olympics.
```

####3.2 Bar Plot

```python
sns.set(style="darkgrid")

sns.barplot(x=df.Sex, y=df.Height, hue=df.Season,palette=['tomato','royalblue'])
plt.ylabel='Height in Centimeter'
plt.title('Barplot: Height by Gender and Season of Olympic Athletes')
plt.show()

#Because the graph looks relatively similar (Summer vs. Winter or M vs. F),
#I don't think the bar plot does a good job at showcasing this data set. 
#It is difficult to make comparisons from the bar plot. 
```

####3.3 Strip Plot

```python
sns.stripplot(x=df.Sex, y=df.Height, hue=df.Season, 
              palette=["olivedrab","palevioletred"], 
              jitter=True, dodge=True,alpha=0.25)
plt.ylabel='Height in Centimeter'
plt.title('Strip Plot: Height by Gender and Season of Olympic Athletes')
plt.show()

#I think the strip plot does a better job than the bar plot
#at showing the range of each category and the difference
#between male vs female as well as the seasons.
```

####3.4 Strips of Observations on top of Violin Plot

```python
sample_df=df.sample(n=200)

ax=sns.violinplot(x=sample_df.Sex, y=sample_df.Height,
                 inner=None, color='0.8')
ax=sns.stripplot(x=sample_df.Sex, y=sample_df.Height, jitter=True, palette=['mediumpurple','salmon'])

plt.ylabel='Height in Centimeter'
plt.title('Strip Plot with Violin Plot: Height by Gender of Olympic Athletes')
plt.show()

#I like that this graph has an interesting visual that makes it stand out,
#the images created are also helpful in determining the frequency and
#range of the data set. 
```

####3.5 Swarms of Observations on top of Box Plot

```python
ax=sns.boxplot(x=sample_df.Sex, y=sample_df.Height,whis=np.inf, palette=['lightseagreen','orange'])
ax=sns.swarmplot(x=sample_df.Sex, y=sample_df.Height, color='0.2')

plt.ylabel='Height in Centimeter'
plt.title('Swarm Plot with Box Plot: Height by Gender of Olympic Athletes')
plt.show()

#I think this graph is helpful in showing the quantiles and also the frequency
#of values within the data set. With its unique visuals, I think it captures
#the attention of the audience who views them. 
```

####3.6 Swarm Plot

```python
ax=sns.swarmplot(x=sample_df.Height,y=sample_df.Sex, palette='Set2')

plt.title('Swarm Plot: Height by Gender of Olympic Athletes')
plt.show()

#Although the visual is interesting, I think the swarm plot with 
#the box plot does a better job at portraying the data set. 
#The swarm plot alone doesn't show the average and other information,
#which makes it difficult for the audience to make comparison.
```
